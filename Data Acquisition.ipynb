```python
import yfinance as yf
```

```python
btc = yf.Ticker('BTC-USD')
prices1 = btc.history(period='8y')
```

```python
prices1
```

```python
usdt = yf.Ticker('USDT-USD')
prices2 = usdt.history(period='5y')
```

```python
data=prices1.join(prices2, lsuffix = ' (BTC)', rsuffix = ' (USDT)')
```

```python
data
```

```python
start_date = '2018-01-01'
end_date = '2022-01-31'
```

```python
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]
```

```python
filtered_data
```

```python
data.to_csv('crypto.csv')
```
