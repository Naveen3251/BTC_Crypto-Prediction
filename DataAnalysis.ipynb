```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('crypto.csv')
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.shape
```

```python
df.set_index('Date',inplace=True)
```

```python
df.info()
```

# Whole data analysis

```python
plt.figure(figsize = (25, 5))
sns.set_style('dark')
sns.lineplot(data=df)
```

# Analysis on Close

```python
plt.figure(figsize = (20, 10))
sns.set_style('dark')
sns.lineplot(data = df['Close (BTC)'], label = 'BTC')
sns.lineplot(data = df['Close (USDT)'], label = 'USDT')
plt.title('Adjacent Close Price')
```

# Analysis on Volume

```python
plt.figure(figsize = (20, 10))
sns.set_style('dark')
sns.lineplot(data = df['Volume (BTC)'], label = 'BTC')
sns.lineplot(data = df['Volume (USDT)'], label = 'USDT')
plt.title('Adjacent Volume Price')
```

# Analysis on Open

```python
plt.figure(figsize = (20, 10))
sns.set_style('dark')
sns.lineplot(data = df['Open (BTC)'], label = 'BTC')
sns.lineplot(data = df['Open (USDT)'], label = 'USDT')
plt.title('Adjacent Open Price')
```

# Analysis on High

```python
plt.figure(figsize = (20, 10))
sns.set_style('dark')
sns.lineplot(data = df['High (BTC)'], label = 'BTC')
sns.lineplot(data = df['High (USDT)'], label = 'USDT')
plt.title('Adjacent High Price')
```

# Analysis on Low

```python
plt.figure(figsize = (20, 10))
sns.set_style('dark')
sns.lineplot(data = df['Low (BTC)'], label = 'BTC')
sns.lineplot(data = df['Low (USDT)'], label = 'USDT')
plt.title('Adjacent Low Price')
```

```python
features_to_investigate = ['Open (BTC)', 'High (BTC)', 'Low (BTC)', 'Close (BTC)', 'Volume (BTC)',
                           'Open (USDT)', 'High (USDT)', 'Low (USDT)', 'Close (USDT)', 'Volume (USDT)']
sns.pairplot(df[features_to_investigate], diag_kind='kde')
```
